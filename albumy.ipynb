```python
import pandas as pd
```

```python
data = pd.read_html('https://www.officialcharts.com/chart-news/the-best-selling-albums-of-all-time-on-the-official-uk-chart__15551/', header=0)
data
```

```python
df=data[0]
df.drop(['POS'],axis=1,inplace=True)
df
```

```python
#Zadanie 1
df.rename(columns={'TITLE':'TYTUŁ','ARTIST':'ARTYSTA','YEAR':'ROK','HIGH POSN':'MAX POZ'},inplace=True)
df
```

```python
#Zadanie 2
df['ARTYSTA'].nunique()
```

```python
#Zadanie 3
print(df['ARTYSTA'].value_counts().head())
```

```python
#Zadanie 4
df.rename(str.title, axis='columns',inplace=True)
df
```

```python
#Zadanie 5
df.drop('Max Poz',axis=1,inplace=True)
df
```

```python
#Zadanie 6
max_count=df['Rok'].value_counts().max()
df['Rok'].value_counts()[df['Rok'].value_counts()==max_count]
```
